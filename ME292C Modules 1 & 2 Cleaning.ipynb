```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
methods = pd.read_csv("methodData.csv")
modules = pd.read_csv("moduleData.csv")
teams = pd.read_csv("teamData.csv")
indAggResponses = pd.read_csv("indAggResponses.csv")
teamAggResponses = pd.read_csv("teamAggResponses.csv")
```

```python
teams292Raw = pd.read_csv("Raw Data/ME 292C Methods and Skills Analysis.csv")
nrows292 = teams292Raw.shape[0]
```

```python
teams292Raw
```

```python
teamDivs = [i for i in range(nrows292) if type(teams292Raw["Team Name/ Mentor"][i]) == str]
teamInds = [range(teamDivs[i-1], teamDivs[i]) for i in range(1, len(teamDivs))]
teamInds.append(range(teamDivs[len(teamDivs) - 1], len(teams292Raw)))
```

```python
teams = [teams292Raw["Team Name/ Mentor"][i] for i in teamDivs]
numTeams = len(teams)
```

```python
teamNames = [team.split('\n')[0] for team in teams]
mentors = [team.split('\n')[1] for team in teams]
```

```python
rMethods = [[teams292Raw["Research Methods Used"][j] for j in teamInds[i] 
             if type(teams292Raw["Research Methods Used"][j]) == str] for i in range(numTeams)]
aMethods = [[teams292Raw["Analysis Methods Used"][j] for j in teamInds[i] 
             if type(teams292Raw["Analysis Methods Used"][j]) == str] for i in range(numTeams)]
iMethods = [[teams292Raw["Ideation Methods Used"][j] for j in teamInds[i] 
             if type(teams292Raw["Ideation Methods Used"][j]) == str] for i in range(numTeams)]
members = [[teams292Raw["Team Members"][j] for j in teamInds[i] 
             if type(teams292Raw["Team Members"][j]) == str] for i in range(numTeams)]
```

```python
methodsChosen = pd.DataFrame({"Team Name": teamNames,
                             "Mentors": mentors,
                             "Research Methods": rMethods,
                             "Analysis Methods": aMethods,
                             "Ideate Methods": iMethods,
                             "Members": members}, 
                             columns=["Team Name", "Research Methods", "Analysis Methods", "Ideate Methods", "Mentors", "Members"])
```

```python
methodsChosen.to_csv("methodsChosen292.csv")
```
